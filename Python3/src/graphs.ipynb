```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Data_sample.txt", sep="|")
```

```python
df2 = df.drop(columns=['   nwhel,steer ','   hding,__mag ','   ruddr,_surf ','   hding,_true '])


pitch_df = df2.drop(columns=['   ailrn,_surf ','   _roll,__deg '])
pitch_df = pitch_df.assign(setpoint=-3)
pitch_df['  _elev,_surf '] = pitch_df['  _elev,_surf '].apply(lambda x: x*10)
roll_df = df2.drop(columns=['  _elev,_surf ','   pitch,__deg '])
roll_df = roll_df.assign(setpoint=0)
roll_df['   ailrn,_surf '] = roll_df['   ailrn,_surf '].apply(lambda x: x*10)
```

```python
sns.lineplot(data=pitch_df)
```

```python
sns.lineplot(data=roll_df)
```

```python
df = pd.read_csv("Oscillations.txt", sep="|")
df3 = df.drop(columns=['   nwhel,steer ','   hding,__mag ','   ruddr,_surf ','   hding,_true '])


pitch_df_2 = df3.drop(columns=['   ailrn,_surf ','   _roll,__deg '])
pitch_df_2 = pitch_df_2.assign(setpoint=0)

roll_df_2 = df3.drop(columns=['  _elev,_surf ','   pitch,__deg '])
roll_df_2 = roll_df_2.assign(setpoint=0)
roll_df_2['   ailrn,_surf '] = roll_df_2['   ailrn,_surf '].apply(lambda x: x*10)
```

```python
sns.lineplot(data=pitch_df_2)
```

```python
sns.lineplot(data=roll_df_2)
```
